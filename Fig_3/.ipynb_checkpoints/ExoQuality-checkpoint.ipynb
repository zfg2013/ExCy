```python
#libraries
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd 
import math
from sklearn.preprocessing import QuantileTransformer
import scipy as scipy
```

```python
print(sns.__version__)
```

```python
#quantile transformation

##data prep
data = pd.read_csv('ExoQuality_Dataset.csv') #import data
new_data = pd.DataFrame(QuantileTransformer(output_distribution='uniform', random_state=389362).fit_transform(data.iloc[:,0:5]), #transform 
                        columns=['PDI','Protein','Yield','RNA','Size'])
new_data['Method']=data['Method'] #add label vector

##graphing time
palette = ['red','gray','orange','cyan'] #set your color palette 
sns.set_context("paper", font_scale=3)
g = sns.pairplot(data=new_data, hue='Method', palette=palette, height=3, aspect=1, plot_kws={"s":80})
for ax in g.axes.flat:
    ax.set_xlabel(ax.get_xlabel(), fontweight='bold')
    ax.set_ylabel(ax.get_ylabel(), fontweight='bold')

#plt.savefig('EQI_Pairplot.jpg')
plt.show()

##sample by standard deviation matrix - ExoQuality

##Dictionary storage
ExCy, FF, EE, UC = [], [], [], []

##Run the loop to iterate through each method in each patient. 
for i in range(11):
    p=4
    p=p*i
    ExCy.append(np.sum((new_data.iloc[0+p:4+p,:5].std())))
    FF.append(np.sum((new_data.iloc[44+p:48+p,:5].std())))
    EE.append(np.sum((new_data.iloc[84+p:88+p,:5].std())))
    UC.append(np.sum((new_data.iloc[132+p:136+p,:5].std())))

    
patient_vector = np.arange(1,12)
    
data = pd.DataFrame({"ExCy":ExCy,
                     "FF":FF,
                     "EE":EE,
                     "UC":UC},
                    index = patient_vector 
)

sns.set_context("paper", font_scale=2.2)
plt.figure(figsize=(13,8))
sns.heatmap(data, cmap="Oranges", annot=True, fmt=".3f", linewidths=0.8, annot_kws={"size":30})
plt.tick_params(axis='both', which='major', labelsize=28, labelbottom = False, bottom=False, top = True, labeltop=True)
plt.xticks(fontweight="bold")
plt.yticks(fontweight="bold")
plt.tight_layout()
#plt.savefig('Standard Deviation heatmap.jpg', dpi=250)
plt.show()
```
