```r
#This block to load all libraries needed
library("DEXSeq")
library("tidyr")
library("qvalue")
library("edgeR")
library("dplyr")
library("reshape2")
library("EnhancedVolcano")
library("sva")
library("Rtsne")
library("ggplot2")
library("pheatmap")
library("org.Hs.eg.db")
library("readr")
```

```r
##GFF
flattenedFile = list.files('.', pattern="gff$", full.names=TRUE)

#METADATA
metadata = read.csv('./ExCy_metadata_dexseq.csv')
rownames(metadata) <- metadata$SampleID
metadata <- metadata[,-1]

#Entire dataset generation here to do downstream analyses
All.countfiles=list.files('./Healthy/', pattern="txt$", full.names=TRUE)

##Generate count file
All.dxd = DEXSeqDataSetFromHTSeq(countfiles = All.countfiles, sampleData = metadata[c(5:12,21:24),], design = ~ sample + exon, flattenedfile = flattenedFile )
All.count <- as.data.frame(featureCounts(All.dxd))

#Do the light filtering and aggregation
All.count$ID <- rownames(All.count)
All.count <- All.count %>% mutate(Gene = substr(ID, 1,15))
All.cf <- All.count %>% mutate(row_sum = rowSums(.[,-13:-14])) %>% dplyr::filter(row_sum > 10) %>% select(-ID)
All.cfs <-  melt(All.cf, id.vars = "Gene") %>% group_by(Gene, variable) %>% summarise(sum_value=sum(value)) %>% dcast(Gene ~ variable, value.var="sum_value")
rownames(All.cfs) <- All.cfs$Gene


#prepare for annotation
exomega <- as.data.frame(read.csv('ExoMEGA.txt'))
mRNA <- read_tsv('VESICLEPEDIA_PROTEIN_MRNAS_5.1.txt')

##map gene names to HUGO
annot.df <- mapIds(org.Hs.eg.db, keys = row.names(All.cfs), column = "SYMBOL", keytype = "ENSEMBL") %>% as.data.frame() %>% rename(SYMBOL=1)
annot.df$SYMBOL <- toupper(annot.df$SYMBOL)
All.cfs$gene_id <- annot.df$SYMBOL 

#filter out non-mapped genes and combine duplicate
All.gcfs <- All.cfs %>% dplyr::filter(!grepl("NA", gene_id, fixed= TRUE)) %>% select(-Gene)
All.gpcfs <-  melt(All.gcfs, id.vars = "gene_id") %>% group_by(gene_id, variable) %>% summarise(sum_value=sum(value)) %>% dcast(gene_id ~ variable, value.var="sum_value")
All.gpcfs <- All.gpcfs[complete.cases(All.gpcfs), ]


All.exomega <- as.data.frame(All.gpcfs[All.gpcfs$gene_id %in% exomega[,1] ,])
rownames(All.exomega) <- All.exomega$gene_id
mRNA_All <- All.exomega[rownames(All.exomega) %in% mRNA$mRNA ,]
All.exomega <- mRNA_All[,2:13]

#write.table(All.exomega, 'raw_counts_healthy.txt', sep="\t")



##########################################################
#edgeR
vol.metadata <- metadata[c(5:12,21:24),]
method <- relevel(factor(vol.metadata$Method), ref="UC")
sample <-factor(vol.metadata$ID)
design <- model.matrix(~0+method)
d <- DGEList(counts = All.exomega)
d <- calcNormFactors(d)
d <- estimateDisp(d, design, robust=TRUE)
d_fit <- glmFit(d, design)


#pairiwse test
qlf_ex_vs_ee <- glmLRT(d_fit, contrast=c(0,1,-1,0))
head(topTags(qlf_ex_vs_ee, n=Inf)$table)

#pairiwse test
qlf_ex_vs_ff <- glmLRT(d_fit, contrast=c(0,1,0,-1))
head(topTags(qlf_ex_vs_ff, n=Inf)$table)

#pairiwse test
qlf_ee_vs_ff <- glmLRT(d_fit, contrast=c(0,0,1,-1))
head(topTags(qlf_ee_vs_ff, n=Inf)$table)


#pairiwse test
qlf_ex_vs_eeandff <- glmLRT(d_fit, contrast=c(0,1,-0.5,-0.5))
head(topTags(qlf_ex_vs_eeandff, n=Inf)$table)


###General test , MAKE SURE TO RESET THE DESIGN TO AN ALL BY ALL OR ELSE THE GLM WILL BE WAY OFF
design <- model.matrix(~method)
d <- DGEList(counts = All.exomega)
d <- calcNormFactors(d)
d <- estimateDisp(d, design, robust=TRUE)
d_fit_gen <- glmFit(d, design)
qlf_gen <- glmLRT(d_fit_gen, coef=2:4) ##we want to find what is different between all 4 groups (n-1 for the test since the df are constrained (if you know 3 then you know the 4th))
head(topTags(qlf_gen, n=Inf)$table)
#write.csv(topTags(qlf, n=Inf)$table, "deg_healthy_mrna.csv")

###############################################################################

#Heatmap
All.heatmap <- All.exomega
Gene_List = rownames(All.exomega)
All.heatmap <- cbind(All.heatmap, Gene_List)
rownames(All.heatmap) <- rownames(All.exomega)

DEG <- topTags(qlf_gen, n=Inf)$table %>% slice(1:103)
All_prep_heatmap <- as.data.frame(All.heatmap[All.heatmap$Gene_List %in% rownames(DEG), ])
All_prep_heatmap <- All_prep_heatmap[,-13] 
colnames(All_prep_heatmap) <- vol.metadata$Method
labels <- colnames(All_prep_heatmap)
            
#graph 
pheatmap(
    log2(All_prep_heatmap+1),
    scale='row',
    border_color="white",
    clustering_method="ward.D2",
    fontsize=16,
    number_color="black",
    display_numbers=F,
    cutree_cols=4,
    show_rownames=F,
    labels_col=labels,
    fontface="bold",
    angle_col=90,
    width=6,
    height=6,
    breaks=seq(-2, 2, length.out=101),
    filename="Healthy_Heatmap_mrna.png"
)

#write.csv(log2(All_prep_heatmap+1), "healthy_deg_mrna_heatmap.csv")



#######################################################

#Rank all the differential transcripts from the heatmap to each isolation method
##Yes. I know, this is some ugly code. 


#restore original object
All_prep_heatmap <- as.data.frame(All.heatmap[All.heatmap$Gene_List %in% rownames(DEG), ])
All_prep_heatmap <- All_prep_heatmap[,-13] 

#rank
Ranked <- log2(All_prep_heatmap+1) %>% rowwise() %>% dplyr::mutate(Ex_avg=mean(c_across(c(1,5,9)))) %>% 
          dplyr::mutate(EE_avg=mean(c_across(c(2,6,10)))) %>%
          dplyr::mutate(FF_avg=mean(c_across(c(3,7,11)))) %>%
          dplyr::mutate(UC_avg=mean(c_across(c(4,8,12)))) %>%
          dplyr::mutate(Ex_rank=case_when(Ex_avg==max(c_across(c('Ex_avg','EE_avg','FF_avg','UC_avg')))~1)) %>%
          dplyr::mutate(EE_rank=case_when(EE_avg==max(c_across(c('Ex_avg','EE_avg','FF_avg','UC_avg')))~1)) %>%
          dplyr::mutate(FF_rank=case_when(FF_avg==max(c_across(c('Ex_avg','EE_avg','FF_avg','UC_avg')))~1)) %>%
          dplyr::mutate(UC_rank=case_when(UC_avg==max(c_across(c('Ex_avg','EE_avg','FF_avg','UC_avg')))~1)) 

Ranked <- cbind(Ranked, Genes=rownames(All_prep_heatmap))

#Totals
ex_to <- sum(complete.cases(Ranked$Ex_rank))
ee_to <- sum(complete.cases(Ranked$EE_rank))
ff_to <- sum(complete.cases(Ranked$FF_rank))
uc_to <- sum(complete.cases(Ranked$UC_rank))


# Create data
data <- data.frame(Methods=c("ExCy","Fujifilm","ExoEasy","UC"), Total=c(ex_to,ff_to, ee_to,uc_to))
 
# Plot
ggplot(data, aes(x=Total, y=Methods)) +
  geom_point() + 
  geom_segment( aes(x=0, xend=Total, y=Methods, yend=Methods)) +
  theme_classic(base_size=25)


#Now extract the top ten ExCy transcripts 
top_10_genes <- arrange(Ranked, desc(Ex_avg)) %>% arrange(., desc(Ex_rank)) %>% slice(1:10) %>% select(Genes)
top_heatmap <- as.data.frame(All.heatmap[All.heatmap$Gene_List %in% top_10_genes$Genes, ]) %>% select(-Gene_List)
prep <- as.data.frame(t(log2(top_heatmap+1))) %>% dplyr::mutate(Methods=c("ExCy","ExoEasy","Fujifilm","UC",
                                                                  "ExCy","ExoEasy","Fujifilm","UC",
                                                                  "ExCy","ExoEasy","Fujifilm","UC")) %>% 
                                          pivot_longer(-Methods, names_to="Genes", values_to="Expression")



##plot the top ten transcripts as a box plot

#color vector
colours <- c("ExCy" ="red3", "ExoEasy"="orange", "Fujifilm"="grey", "UC"="blue2")
#plotting
options(repr.plot.width = 12, repr.plot.height =10)

ggplot(prep, aes(x=Expression, y=Genes, fill=Methods)) + 
  geom_boxplot()+
  scale_fill_manual(values=colours)+
  geom_point(position=position_dodge(width=0.75), aes(group=Methods))+
  labs(x = "Log2 mRNA Transcripts", y = "Top 10 ExCy Transcripts") +
  theme_classic() +
    theme(axis.line = element_line(color='black'),
    plot.background = element_blank(),
    panel.grid.major = element_blank(),
    panel.grid.minor = element_blank(),
    axis.text.x = element_text(size=25, face="bold", color="black"),
    axis.text.y = element_text(size=25, color="black"),
    text = element_text(size=25, face='bold')
          )



###################################################
```
